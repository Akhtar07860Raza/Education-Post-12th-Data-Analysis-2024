# Remove outliers from quantitative columns

```python
def remove_outliers(df, columns):
  for col in columns:
    # Calculate the interquartile range (IQR)
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    # Define the lower and upper bounds for outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Remove outliers
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

  return df
```

```python
# Remove outliers from quantitative columns
data = remove_outliers(data, quantitative_cols)

# Create a single boxplot graph for all quantitative columns
plt.figure(figsize=(15, 5))
sns.boxplot(data=data[quantitative_cols])
plt.title('Boxplots of all quantitative columns after removing outliers')
plt.xticks(rotation=90)
plt.show()
```
